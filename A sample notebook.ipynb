```python
power = 7
```

```python
import matplotlib.pyplot as plt

ys = [x**power for x in range(-100,100)]

plt.scatter(range(len(ys)), ys)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("A simple function")
plt.show()
```
